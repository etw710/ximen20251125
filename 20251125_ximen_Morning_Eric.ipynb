````python
# calculate bmi: weight(kg) / height(m) ^ 2
weight = float(input("Please enter your weight(kg): "))
height = float(input("Please enter your height(cm): "))
bmi = weight / (height / 100) ** 2

print("Your weight is: " + str(weight))
print("Your height is: " + str(height))
print("Your calculated BMI is: " + str(bmi))

# 縮排 需要 ":"冒號 "TAB"縮排 indentationError
# 過濾單選題: 嚴格放上面
if bmi >= 25:
    print("over")
elif bmi >= 18: # 是單選題, 不需要打 25 > bmi >= 18, 因為前面 if 條件就已經達成了
    print("normal")
else:
    print("under")
# print("!!!!") # 如果縮在前面, 每一個狀況都會出現
````
